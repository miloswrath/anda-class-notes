```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
import numpy as np
```

```python
# Load the iris dataset
iris = pd.read_csv('iris.csv')
```

```python
# Show head
iris.head()
```

```python
# print tail (last 8)
iris.tail(n=8)
```

```python
# Print shape
print(f"Shape in rows and columns {iris.shape}")
```

```python
# describe dataset
iris.describe()
```

```python
# Print info (dtypes, missing values)
iris.info()
```

```python
# Print number of unique values in 'species' column
print(f"Number of unique values in 'species' column: {len(iris['species'].unique())}")
```

```python
# print number of unique values in 'sepal_length' column
print(f"Number of unique values in 'sepal_length' column: {len(iris['sepal_length'].unique())}")
```

```python
# load seaborn and plot pairplot
import seaborn as sns
sns.pairplot(iris, hue='species')
```

```python
# print correlation matrix between all numeric columns
numeric_columns = iris.select_dtypes(include=[np.number])
correlation_matrix = numeric_columns.corr()
correlation_matrix
```

```python
# import matplotlib and plot scatterplot between petal_length and petal_width
import matplotlib.pyplot as plt
colors = iris.species.map({'setosa': 'green', 'versicolor': 'blue', 'virginica': 'brown'})
iris.plot(kind='scatter', x='petal_width', y='petal_length', c=colors)
```
